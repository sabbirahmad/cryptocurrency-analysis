## Cryptocurrency


```python
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```python
def load_dataset(filename):

    with open(filename,'r') as dest_f:
        data_iter = csv.reader(dest_f, delimiter = ',', quotechar = '"')
        data = [data for data in data_iter]
        data_array = np.asarray(data)
        
    return data_array

def load_pd_dataset(filename):
    data = pd.read_csv(filename)
    return data
```

```python
dataset = load_pd_dataset('crypto-markets.csv')
```

```python
d = dataset[dataset['ranknow'] == 1]
d = d[['open','high', 'low', 'close']]

fig = plt.figure(figsize=(4, 4))
box = d.boxplot(grid=True, return_type='axes', showfliers=False)
plt.show()
```

```python
scaler = MinMaxScaler()
dat = dataset[['open','high', 'low', 'close']]
scaler.fit(dat)
dat = scaler.transform(dat)
dat
```

```python
fig = plt.figure(figsize=(18, 20))
sub = []
ncols = 4
nrows = 5
ind = 1
for rank in range(1, 9):
    sub.append(plt.subplot(nrows, ncols, ind))
    sub[ind-1].set_title(label='rank = '+ str(rank), fontsize=12)
    d = dataset[dataset['ranknow'] == rank]
    d = d[['open','high', 'low', 'close']]
    box = d.boxplot(grid=True, return_type='axes', showfliers=False)
    ind += 1

fig.tight_layout()
plt.show()
```

```python
fig = plt.figure(figsize=(18, 20))
sub = []
ncols = 4
nrows = 5
ind = 1
for rank in range(1, 9):
    sub.append(plt.subplot(nrows, ncols, ind))
    sub[ind-1].set_title(label='rank = '+ str(rank), fontsize=12)
    d = dataset[dataset['ranknow'] == rank]
    d = d[['open','high', 'low', 'close']]
    scaler.fit(d)
    d = scaler.transform(d)
    d = pd.DataFrame(d, columns=['open','high', 'low', 'close'])
    
    box = d.boxplot(grid=True, return_type='axes', showfliers=False)
    ind += 1

fig.tight_layout()
plt.show()
```
