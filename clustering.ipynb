```python
import pandas as pd
import numpy as np
```

```python
fileLoc = './all-crypto-currencies/crypto-markets.csv'
data = pd.read_csv(fileLoc)
data.dropna()
```

```python
unique_coins = data.name.unique()
bit_coin = data.loc[data['symbol'] == 'BTC']

coins_last = data.loc[data['date'] == '2018-03-10']
coins_last
```

```python
coins_last = coins_last.reset_index(drop='true')
```

```python
coins_table = coins_last.drop(columns=['index', 'slug', 'symbol', 'name', 'ranknow','date'])
coins_table
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

plt.rcParams['figure.figsize'] = (16, 9)

# Creating a sample dataset with 4 clusters
X, y = make_blobs(n_samples=800, n_features=3, centers=4)
```

```python
# Initializing KMeans
kmeans = KMeans(n_clusters=4)
# Fitting with inputs
kmeans = kmeans.fit(coins_table)
# Predicting the clusters
labels = kmeans.predict(coins_table)
# Getting the cluster centers
C = kmeans.cluster_centers_
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y)
ax.scatter(C[:, 0], C[:, 1], C[:, 2], marker='*', c='#050505', s=1000)
```
